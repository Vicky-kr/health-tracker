```python
#importing necessary packages
import cv2
from PIL import Image
import requests
from io import BytesIO
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy
from PIL import Image
from numpy import array
import skimage.io
#Loading the face detector model
face_cascade = cv2.CascadeClassifier('face_detector.xml')
```

```python

###Reading and storing as numpy array#Initializing instances
```

```python
#Defining the class
class Face_Data:
    def __init__(self,name,loc,pict):
        self.name=name                
        self.loc=loc
        self.pict=pict
    
    
    #These three numpy arrays are to be stored in database
    
    #Draw rectangle around the faces
    def marking_faces(faces,ar1):
        for (x, y, w, h) in faces:
            cv2.rectangle(ar1, (x, y), (x+w, y+h), (255, 0, 0), 2)
        img = skimage.io.imread('face_detected.png')
        cv2.imwrite('sample_out_2.png',img) 
        #cv2.imwrite('face_detected.png', ar1) 
    def valid_photo_check(self,img):
        ar1=array(img) # COnvert image to a matrix
        faces = face_cascade.detectMultiScale(ar1, 1.1, 4)
        no_of_faces_detected=len(faces)
        if no_of_faces_detected==1:
            print('SUCCESSFULLY SAVED! WELCOME TO HEALTHPLUS!!!')
        elif no_of_faces_detected==0:
            print('WARNING!NO FACES DETECTED! Please make sure you are in a well lit place')
        else:
            print('MULTIPLE FACES DETECTED!! PLEASE TRY AGAIN')
        
        for (x, y, w, h) in faces:
            cv2.rectangle(ar1, (x, y), (x+w, y+h), (255, 0, 0), 2)
        #arr = skimage.io.imread(img)
        cv2.imwrite('sample_out_2.png',ar1) 
        imag_final = cv2.imread('sample_out_2.png')
        imgplot = plt.imshow(imag_final)
        plt.show()
        print('NOTE:ONLY HUMAN FACE(S) ARE MARKED. MULTIPLE FACES/PRESENCE OF ANIMALS/ OBJECTS RESEMBLING FACES MAY DEEM THE IMAGE INVALID')

    

#sample_out_2: Saved image in original colors
#face_detected: Saved image in bgr format
    def final_message():
        imag_final = mpimg.imread('sample_out_2.png')
        imgplot = plt.imshow(imag_final)
        plt.show()
        print('NOTE:ONLY HUMAN FACE(S) ARE MARKED. MULTIPLE FACES/PRESENCE OF ANIMALS/ OBJECTS RESEMBLING FACES MAY DEEM THE IMAGE INVALID')
```
